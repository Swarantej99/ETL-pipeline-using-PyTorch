```python
import torch
from torch.utils.data import Dataset, DataLoader
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
file_path = "Users/allullaswarantej/Downloads/temperature.csv"
df = pd.read_csv(file_path)
print("Dataset loaded successfully. First few rows:")
print(df.head())
```

```python
# Drop non-numeric columns
df_numeric = df.iloc[:, 4:]  # Exclude first four columns (metadata)
```

```python
# Handle missing values (fill with mean)
df_numeric.fillna(df_numeric.mean(), inplace=True)
```

```python
# Normalize data
scaler = MinMaxScaler()
df_normalized = scaler.fit_transform(df_numeric)

def tensor_conversion(data):
    return torch.tensor(data, dtype=torch.float32)
```

```python
class TemperatureDataset(Dataset):
    def __init__(self, data):
        self.data = tensor_conversion(data)
    
    def __len__(self):
        return len(self.data)
    
    def __getitem__(self, idx):
        return self.data[idx]

# Create dataset and DataLoader
dataset = TemperatureDataset(df_normalized)
dataloader = DataLoader(dataset, batch_size=16, shuffle=True)
```

```python
for batch in dataloader:
    print("Sample batch:", batch)
    break
```
